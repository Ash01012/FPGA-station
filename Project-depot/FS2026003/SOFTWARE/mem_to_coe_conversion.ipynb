```python

def mem_to_coe(input_path, output_path, radix=2):
    with open(input_path, "r") as f:
        lines = [
            line.strip()
            for line in f
              if line.strip() and not line.strip().startswith("//")
        ]

    with open(output_path, "w") as f:
        f.write(f"memory_initialization_radix={radix};\n")
        f.write("memory_initialization_vector=\n")

        for i, data in enumerate(lines):
            terminator = ";" if i == len(lines) - 1 else ","
            f.write(f"{data}{terminator}\n")


mem_to_coe("Xtest_6.mem", "Xtest_6.coe", radix=2)
mem_to_coe("Y_test.mem", "Y_test.coe", radix=2)
mem_to_coe("clauses.mem", "clauses.coe", radix=2)
mem_to_coe("Xtest_160_2.mem", "Xtest_160_2.coe", radix=2)
mem_to_coe("Ytest_160_2.mem", "Ytest_160_2.coe", radix=2)
mem_to_coe("clauses_160_2.mem", "clauses_160_2.coe", radix=2)
mem_to_coe("weights_160_2.mem", "weights_160_2.coe", radix=2)
```
